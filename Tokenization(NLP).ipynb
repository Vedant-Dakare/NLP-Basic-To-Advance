```python
corpus = """ Hello welcome to my journey of machine learning.
Now i started NLP . Previously it seems horrible!! but now it is okay.
"""
```

```python
print(corpus)
```

```python
## Tokenization
## paragraph --> sentence
from nltk.tokenize import sent_tokenize
sent_tokenize(corpus)
```

## stemming
- stemming means reducing a word to its word stem

```python
## classification of the comments in the postive or negtive
## words --> eaten eats eating --> stem -> eat
words = ["eating","eats","eaten","programming","programs","history","finally","finalize"]
```

## Porter stemmer

```python
from nltk.stem import PorterStemmer
```

```python
steming = PorterStemmer()
```

```python
for word in words:
  print(word + "---->" + steming.stem(word))
```

```python
stemming.stem("congratulations")
```

```python
stemming.stem("sitting")
```

## RegexpStemmer

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$',min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

## Snowball stemmer

```python
from nltk.stem import SnowballStemmer
```

```python
snowballstemmer=SnowballStemmer('english')
```

```python
for word in words:
  print(word + '---->' + snowballstemmer.stem(word))
```

## why snowball stemmer is better that porter and all stemming techniques

```python
## porter stemming
stemming.stem("fairly"),stemming.stem("sportingly")
```

```python
## use snowball stemmer
snowballstemmer.stem("fairly"),snowballstemmer.stem("sportingly")
```

## for many words snowball stemmer give good output so mostly use snowball stemmer

```python
stemming.stem("goes")
```

```python
snowballstemmer.stem("goes")
```

## FOR SOME OF THE WORD STEMMING IS NOT WORKING FINE SO WE USE LEMMATIZATION

## LEMMATIZATION
- by the lemmatization technique we get the root word
- output of the lemmatization is "lemma"
- use mostly in Q&A , chatbots, text summarization

```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
'''
POS - Noun - n
verb - v
adjective - a
adverb - r
'''
lemmatizer.lemmatize("going",pos='v')
```

```python
for word in words:
  print(word + '---->' + lemmatizer.lemmatize(word,pos='v'))
```

```python
lemmatizer.lemmatize("fairly",pos='v'),lemmatizer.lemmatize("sportingly",pos='v')
```

## lemmatization is good with respect to stemming but it takes more time with respect to stemming when we have large copus(paragraph)
